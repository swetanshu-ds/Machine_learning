```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("wine_fraud.csv")
```

```python
df.drop(columns = ['type'],inplace =True)
```

```python
x = df.drop(columns = "quality")
```

```python
y = df.quality
```

```python
train,val_train,test,val_test = train_test_split(x,y,test_size = 0.50,random_state = 42)
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,test,test_size = 0.3,random_state = 42)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
prediction_kn = knn.predict(val_train)
```

```python
prediction_svc = svc.predict(val_train)
```

```python
input1 = np.column_stack((prediction_kn,prediction_svc))
```
